```python
# Creating a dataframe from a dict

test_df = pd.DataFrame( { 'col1' : ['value1'], 'col2' : ['value2'], 'col3' : ['value3'] } ) ## Note: must be a dict of lists, even if the column contains only a single value
```

```python
# Creating a dataframe from a list of lists 
# ( one list per row )

test_df2 = pd.DataFrame( [ ['r1c1', 'r1c2', 'r1c3'],\
                        ['r2c1', 'r2c2', 'r2c3']] ) # without columns param, will name columns from zero index

test_df3 = pd.DataFrame( [ ['r1c1', 'r1c2', 'r1c3'],\
                        ['r2c1', 'r2c2', 'r2c3']] \
                       ,columns = [ 'column1', 'column2', 'column3' ]) 
```

```python
print( test_df)
print( test_df2)
print( test_df3)
```
